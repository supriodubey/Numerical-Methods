## The Shooting Method

```python
import numpy as np
import matplotlib.pyplot as plt
import math
N= 10**3
t_int = 0
t_fin = 3
tolerance = 10**(-3)
x_fin= 10
x_in = 0
h = (t_fin-t_int)/N
x =np.zeros(N)
t =np.zeros(N)
v = np.zeros(N)
x[0]= 0.
v[0] = 1.
pos_final = 1.
b=0
while (abs(pos_final-x_fin)> tolerance):

    if pos_final < x_fin:
        v[0]+=0.001
    else:
        v[0]-=0.001

    for i in range(1,N):
            x[i] = x[i-1] + v[i-1]*h
            v[i] = v[i-1] -9.81*h
            t[i] = t[i-1]+h

    pos_final = x[N-1]
    b+=1
plt.plot(t,x)
plt.plot(t,v)
    
```
