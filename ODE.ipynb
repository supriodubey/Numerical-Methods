```python
import numpy as np
import matplotlib.pyplot as plt


def func(x,t):
    f = -x**3 + np.sin(t)
    return f
```

## Euler Method

```python
def euler(x_0,t_int,t_fin,h):
    N = int((t_fin-t_int)/h)
    x = np.zeros(N)
    t = np.zeros(N)
    

    for i in range(N):
        
        t[i]= t[i-1]+h
        
        x[i] = x[i-1] + h*func(x[i-1],t[i-1])

        
    return x ,t

values = euler(0.,0.,100.,0.4) 
# print(values[0])
# print(values[1])
plt.plot(values[1],values[0])    
```

## Midpoint Method(Second Order Runge-Kutta)

```python
def midpoint(x_0,t_int,t_fin,h):
    N = int((t_fin-t_int)/h)
    x = np.zeros(N)
    t = np.zeros(N)
    
    for i in range(1,N):
        k_1 = (h/2)*func(x[i-1],t[i-1])
        k_2 = h*func(x[i-1]+ k_1,t[i-1]+h/2)
        t[i] = t[i-1]+h
        x[i] = x[i-1] + k_2


        
    return x ,t
mid = midpoint(0.,0.,100.,0.4) 
# print(mid[0])
# print(mid[1])
plt.plot(values[1],values[0],label = 'Euler')
plt.plot(mid[1],mid[0],label =  'Midpoint') 

plt.xlabel('X')
plt.ylabel('t')
plt.legend()
```

```python
def rungekutta(x_0,t_int,t_fin,h):
   
    N = int((t_fin-t_int)/h)
    x = np.zeros(N)
    t = np.zeros(N)
    
    for i in range(1,N):
        k_1 = (h/2)*func(x[i-1],t[i-1])
        k_2 = (h/2)*func(x[i-1]+ k_1,t[i-1]+h/2)
        k_3 = h*func(x[i-1]+ k_2,t[i-1]+h/2)
        k_4 = h*func(x[i-1]+ k_3,t[i-1]+h)
        t[i] = t[i-1]+h
        x[i] = x[i-1] + (1/6)*(2*k_1 + 4*k_2 + 2*k_3 + k_4)
        
        return x,t

runge = rungekutta(0.,0.,100.,0.4)
################# Plot ##########
plt.plot(values[1],values[0],label = 'Euler')
plt.plot(mid[1],mid[0],label =  'Midpoint') 
plt.plot(runge[1],runge[0],label = 'Runge-Kutta 4')
plt.legend()
```
