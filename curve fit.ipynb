## Fitting

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize as opt


def data_set():
    a=0.
    s=1.
    h=10.
    x=np.linspace(-10.,10.,num=500) #x data
    y=h*np.exp(-(x-a)**2/2./s**2)
    sigma=np.zeros(len(x),float)
    for i in range(len(y)):
        sigma[i]=4.*np.random.rand()
        y[i]=np.random.normal(y[i],sigma[i]) #add gaussian noise
    return x,y
x,y = data_set()
plt.scatter(x,y,marker = '*')
```

```python
a=0.
s=1.
h=10.
def func(x,a,s,h):
    y=h*np.exp(-(x-a)**2/2./s**2)
    
    return(y)
    

popt,pcov = opt.curve_fit(func,x,y,p0 =(a,s,h))
print(popt)

ynew = func(x,*popt)
plt.scatter(x,ynew)
plt.scatter(x,y,marker = '*')
```
